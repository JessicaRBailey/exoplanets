```python
soup = BeautifulSoup(html, 'html.parser')
```

```python
all_numbers = soup.find_all('span', {'class' : 'animated_number'})
all_numbers = [metric.text for metric in all_numbers]
all_numbers
```

```python
all_titles = soup.find_all('div', {'class' : 'title visually_impaired'})
all_titles = [title.text for title in all_titles]
all_titles
```

```python
exoplanet_metrics = {}

for key, value in zip(all_titles, all_numbers):
    if key not in exoplanet_metrics:
        exoplanet_metrics[key] = value
        
exoplanet_metrics
```

```python
# Establish connection to MongoDB
client = pymongo.MongoClient("mongodb://localhost:27017/")
database = client["ExoplanetDB"]
collection = database["exoplanet_counts"]

data_to_insert = [
    {"key": key, "value": value}
    for key, value in exoplanet_metrics.items()
]

collection.insert_many(data_to_insert)
```
