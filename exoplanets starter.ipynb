```python
import requests
import pandas as pd
import math
```

```python
# API URL
api_url = "https://exoplanetarchive.ipac.caltech.edu/TAP/sync?query=select+*+from+pscomppars&format=json"
```

```python
# Make get request to the API
response = requests.get(api_url)
```

```python
# Check if the request was successful (status code 200?)
Planet_Name = []
Discovered_Year = []
Discovery_Method = []
Orbital_Period_days = []
SemiMajor_Axis_AU = []
Mass = []
Stellar_Luminosity = []
ra = []
rastr = []
dec = []
decstr = []
glon = []
glat = []
elon = []
elat = []
if response.status_code == 200:
    # Parse JSON response
    data = response.json()
    # Test to print the first few records
    for entry in data:
        Planet_Name.append(entry.get("pl_name"))
        Discovered_Year.append(entry.get("disc_year"))
        Discovery_Method.append(entry.get("discoverymethod"))
        Orbital_Period_days.append(entry.get("pl_orbper"))
        SemiMajor_Axis_AU.append(entry.get("pl_orbsmax"))
        Mass.append(entry.get("pl_bmassj"))
        Stellar_Luminosity.append(entry.get("st_lum")) 
        ra.append(entry.get("ra")) 
        rastr.append(entry.get("rastr")) 
        dec.append(entry.get("dec")) 
        decstr.append(entry.get("decstr")) 
        glon.append(entry.get("glon")) 
        glat.append(entry.get("glat")) 
        elon.append(entry.get("elon")) 
        elat.append(entry.get("elat")) 
else:
    print("Error:", response.status_code)
```

```python
exoplanet_data = pd.DataFrame({'Planet Name':Planet_Name,'Discovery Year':Discovered_Year,'Discovery Method':Discovery_Method,'Orbital Period (days)':Orbital_Period_days,'SemiMajor Axis (AU)':SemiMajor_Axis_AU,'Mass (Jupiter Mass)':Mass,'Stellar_Luminosity':Stellar_Luminosity,'Right Ascension (RA)':ra,'Declination (Dec)':dec,'RASTR':rastr,'DECSTR':decstr,'GLON':glon,'GLAT':glat,'ELON':elon,'ELAT':elat })
exoplanet_data = exoplanet_data.dropna()
print(len(exoplanet_data))
exoplanet_data.head(10)
```

```python
csv_filename = 'exoplanet_data.csv'
exoplanet_data.to_csv(csv_filename, index=False)

print(f'Data saved to {csv_filename}')
```

```python
lum = exoplanet_data.Stellar_Luminosity
divisor_ri = 1.1
exoplanet_data['Ri (AU)'] = [math.sqrt(abs(x/divisor_ri)) for x in lum]
divisor_ro = 0.53
exoplanet_data['Ro (AU)'] = [math.sqrt(abs(x/divisor_ro)) for x in lum]
exoplanet_data['Habitable Zone'] = ['Yes' if exoplanet_data['Ri (AU)']>0.95 and exoplanet_data['Ri (AU)']<1.37 else 'No']
exoplanet_data.head()
```
